```python
#
#  RUN THIS SCRIPT FOR GENERATING RANDOM DATA
#

import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
import pickle

# I am going to create 4 ranges of prices, from
# 150k to 200k
# 200k to 225k
# 225k to 250k
# 250 to 500k

df1 = pd.DataFrame()
df1['price'] = np.random.randint(150000, 180000, size=500)
df1['number_of_rooms'] = 2
df1['is_city'] = 0
df1['good_area'] = 0
df1['surface'] = np.random.randint(40, 50, size=500)
df1['rating'] = np.random.randint(0, 1, size=500)
df1['unemployment'] = np.random.randint(10, 15, size=500)
df1['parking'] = 0

df2 = pd.DataFrame()
df2['price'] = np.random.randint(180000, 190000, size=500)
df2['number_of_rooms'] = 3
df2['is_city'] = 0
df2['good_area'] = 0
df2['surface'] = np.random.randint(50, 60, size=500)
df2['rating'] = np.random.randint(1, 2, size=500)
df2['unemployment'] = np.random.randint(8, 10, size=500)
df2['parking'] = 0

df3 = pd.DataFrame()
df3['price'] = np.random.randint(190000,200000, size=500)
df3['number_of_rooms'] = 3
df3['is_city'] = 1
df3['good_area'] = 1
df3['surface'] = np.random.randint(60, 70, size=500)
df3['rating'] = np.random.randint(2, 3, size=500)
df3['unemployment'] = np.random.randint(6, 8, size=500)
df3['parking'] = 1

df4 = pd.DataFrame()
df4['price'] = np.random.randint(200000, 210000, size=500)
df4['number_of_rooms'] = 4
df4['is_city'] = 1
df4['good_area'] = 1
df4['surface'] = np.random.randint(70, 80, size=500)
df4['rating'] = np.random.randint(3, 4, size=500)
df4['unemployment'] = np.random.randint(4, 6, size=500)
df4['parking'] = 1

df = pd.concat([df1, df2, df3, df4])
df.reset_index(drop=True)
df.to_csv("training_df.csv", index=False)
training_df = pd.read_csv("training_df.csv")

scaler = MinMaxScaler(feature_range=(0, 1))
scaling_training_df = scaler.fit_transform(training_df)
scaled_training_df = pd.DataFrame(scaling_training_df, columns=training_df.columns.values)
print("Price has been multiply by {} and add {}".format(scaler.scale_[0], scaler.min_[0]))
scalers = pd.DataFrame({"a": scaler.scale_[1], "b": scaler.min_[0]}, index=[0])

# To CSV
scaled_training_df.to_csv("training_df_scaled.csv", index=False)
scalers.to_csv('scalers.csv', index=False)

# Creating test
df5 = pd.DataFrame()
df5['price'] = np.random.randint(205000, 210000, size=1)
df5['number_of_rooms'] = 4
df5['is_city'] = 1
df5['good_area'] = 1
df5['surface'] = 75
df5['rating'] = 3.5
df5['unemployment'] = 4
df5['parking'] = 1

df5.to_csv('new_house.csv', index=False)
simple_test_df = pd.read_csv('new_house.csv')

scaling_simple_test_df = scaler.transform(simple_test_df)
scaled_simple_test_df = pd.DataFrame(scaling_simple_test_df, columns=simple_test_df.columns.values)

scaled_simple_test_df.to_csv('new_house_scaled.csv', index=False)


with open("scaler.pkl", "wb") as f:
    pickle.dump(scaler, f)
```

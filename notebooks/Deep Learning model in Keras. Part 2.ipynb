```python
import pandas as pd
from keras.models import Sequential
from keras.layers import Dense
from sklearn.model_selection import train_test_split

# Defining Model
model = Sequential()
model.add(Dense(60, input_dim=7, activation='relu'))
model.add(Dense(120, activation='relu'))
model.add(Dense(60, activation='relu'))
model.add(Dense(1, activation='linear'))
model.compile(loss='mean_squared_error', optimizer='adam')

# Loading data from csv and separating randomly 20% for testing
data_df = pd.read_csv("training_df_scaled.csv")
training_df, test_df = train_test_split(data_df, test_size=0.2)

# Separating features and labels
X = training_df.drop('price', axis=1).values
Y = training_df[['price']].values

# Fitting the model
model.fit(
    X,
    Y,
    epochs=40,
    shuffle=True,
    verbose=2
)

# Testing the model
X_test = test_df.drop('price', axis=1).values
Y_test = test_df[['price']].values

# Evaluating the model
error = model.evaluate(X_test, Y_test, verbose=0)
print("Mean Squared Error: {}".format(error))

# Saving the model
model.save("trained_model.h5")
print("Model saved.")

# Making a single prediction
new_house_df = pd.read_csv("new_house_scaled.csv")

X_new = new_house_df.drop('price', axis=1).values
Y_new = new_house_df[['price']].values

prediction = model.predict(X_new)
```
